## Bài giải

**Câu 1: Tính z và $\hat{y}$**  

Ta có z= wx + b = 0.5*2 - 0.4 = 0.6, do $\hat{y}$ = z = 0.6  

**Câu 2: Tính loss**  

Ta có $L = \frac{1}{2}(\hat{y} - y)^2 = \frac{1}{2}(0.6 - 1.2)^2 = 0.18$  

**Câu 3: Tính các gradient**  

* ## <u><i>3.1) Tính gradient $\frac{\partial L}{\partial w}$</i></u>

Theo đề bài ta có w,x,b -> z -> y -> L  Nên $\quad \quad \frac{\partial L}{\partial w} = \frac{\partial L}{\partial \hat{y}} \times \frac{\partial \hat{y}}{\partial z} \times \frac{\partial z}{\partial w}$  

- ### 3.1.1) Tìm $\frac{\partial L}{\partial \hat{y}}$  

Ta có: $L = \frac{1}{2}(\hat{y} - y)^2$  
Đặt u = $\hat{y} - y$  từ đó ta có $\hat{y} \rightarrow u \rightarrow L$  Nên

$\frac{\partial L}{\partial \hat{y}}$ = $\frac{\partial L}{\partial u} \times \frac{\partial u}{\partial \hat{y}}$ = $\frac{\partial}{\partial u}(\frac{1}{2}u^2) \times \frac{\partial (\hat{y} - y)}{\partial \hat{y}}$ = u $\times (\frac{\partial \hat{y}}{\partial \hat{y}} - \frac{\partial y}{\partial \hat{y}})$ = u x 1 = u = $\hat{y}$ - y = 0.6 - 1.2 = -0.6  

Vì theo công thức đạo hàm thì: 

$\frac{\partial y}{\partial \hat{y}} = 0$ -  do y là hằng số  

$\frac{\partial \hat{y}}{\partial \hat{y}} = 1$  

$\frac{\partial}{\partial u}(\frac{1}{2}u^2) = u$  

- ### 3.1.2) Tìm $\frac{\partial \hat{y}}{\partial z}$   

Theo đề bài ta có z = $\hat{y}$ nên $\frac{\partial \hat{y}}{\partial z}$  = $\frac{\partial z}{\partial z}$  = 1  

- ### 3.1.3) Tìm $\frac{\partial z}{\partial w}$   

Theo đề bài ta có: z=wx + b , w là biến, x và b là hằng số 
$\frac{\partial z}{\partial w}$ = $\frac{\partial (wx + b)}{\partial w}$ = $\frac{\partial wx}{w} + \frac{\partial b}{\partial w}$ = x + 0 = 2 + 0 = 2  
Vì: Theo công thức đạo hàm thì $\frac{\partial wx}{w}$ = x và $\frac{\partial b}{\partial w}$ =  0 (do tử chỉ có hằng số là b)  

Từ đó ta có: $\frac{\partial L}{\partial w}$ = $\frac{\partial L}{\partial \hat{y}} \times \frac{\partial \hat{y}}{\partial z} \times \frac{\partial z}{\partial w}$  = -0.6 * 1 * 2 = -1.2 

* ## <u><i>3.2) Tính gradient $\frac{\partial L}{\partial b}$</i></u>  

Theo đề bài ta có w, x, b -> z -> $\hat{y}$ -> L Nên  $\frac{\partial L}{\partial b}$ = $\frac{\partial L}{\partial \hat{y}} \times \frac{\partial \hat{y}}{\partial z} \times \frac{\partial z}{\partial b}$  

- ### 3.2.1) Tìm $\frac{\partial L}{\partial \hat{y}}$  

Ta có: $L = \frac{1}{2}(\hat{y} - y)^2$  
Đặt u = $\hat{y} - y$  từ đó ta có $\hat{y} \rightarrow u \rightarrow L$  Nên

$\frac{\partial L}{\partial \hat{y}}$ = $\frac{\partial L}{\partial u} \times \frac{\partial u}{\partial \hat{y}}$ = $\frac{\partial}{\partial u}(\frac{1}{2}u^2) \times \frac{\partial (\hat{y} - y)}{\partial \hat{y}}$ = u $\times (\frac{\partial \hat{y}}{\partial \hat{y}} - \frac{\partial y}{\partial \hat{y}})$ = u x 1 = u = $\hat{y}$ - y = 0.6 - 1.2 = -0.6  

Vì theo công thức đạo hàm thì: 

$\frac{\partial y}{\partial \hat{y}} = 0$ -  do y là hằng số  

$\frac{\partial \hat{y}}{\partial \hat{y}} = 1$  

$\frac{\partial}{\partial u}(\frac{1}{2}u^2) = u$  


- ### 3.2.2) Tìm $\frac{\partial \hat{y}}{\partial z}$   

Theo đề bài ta có z = $\hat{y}$ nên $\frac{\partial \hat{y}}{\partial z}$  = $\frac{\partial z}{\partial z}$  = 1  

- ### 3.2.3) Tìm $\frac{\partial z}{\partial b}$   

Theo đề bài ta có: z=wx + b , b là biến, w và x là hằng số 
$\frac{\partial z}{\partial w}$ = $\frac{\partial (wx + b)}{\partial w}$ = $\frac{\partial wx}{w} + \frac{\partial b}{\partial w}$ = 1 + 0 = 1
Vì: Theo công thức đạo hàm thì $\frac{\partial wx}{w}$ = 0 (do tử chỉ có hằng số là wx) và $\frac{\partial b}{\partial b}$ =  1   

Từ đó ta có: $\frac{\partial L}{\partial b}$ = $\frac{\partial L}{\partial \hat{y}} \times \frac{\partial \hat{y}}{\partial z} \times \frac{\partial z}{\partial b}$ = -0.6 * 1 * 1 = -0.6
