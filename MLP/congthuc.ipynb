# Bắt đầu

* Mọi vector được viết dưới dạng vector cột.
* Làm tròn đến 4 chữ số thập phân khi cần.  

## Ký hiệu thống nhất  

* Sinh viên phải ghi đầy đủ các biến trung gian: $\quad\quad\quad\quad z^{(l)}, a^{(l)}, b^{(l)}$  

* Với một mẫu dữ liệu x, dùng quy ước: $\quad\quad\quad\quad z^{(l)} = W^{(l)}a^{(l-1)} + b^{(l)}, \quad a^{(l)} = f^{(l)}(z^{(l)}), \quad a^{(0)} = x$  

* Với MSE: $\quad\quad\quad\quad L = \frac{1}{2}(\hat{y} - y)^2$  

* Với softmax + cross-entropy: $\quad\quad\quad\quad L = - \sum_{k=1}^{K} y_k \ln \hat{y_k}$  

* Với một đầu ra vô hướng: $\quad\quad\quad\quad \hat{y} = a^{(L)}, \quad L = \frac{1}{2} (\hat{y} - y)^2$  

* Backprop:  $\quad\quad\quad\quad \delta^{(L)} = \frac{\partial L}{\partial z^{(L)}},$ $\quad \delta^{(L)} = (W^{l+1})^T\delta^{l+1} \odot f'(z^{(l)})$  

* Gradient:  $\quad\quad\quad\quad \frac{\partial L}{\partial W^{(l)}} = \delta^{(l)}(a^{(l-1)})^T, \quad \frac{\partial L}{\partial b^{(l)}} = \delta^{(l)}$  

* Qui tắc cập nhật:  $\quad\quad\quad\quad W^{(l)} \leftarrow W^{(l)} - \eta \frac{\partial L}{\partial W^{(l)}}, \quad b^{(l)} \leftarrow b^{(l)} - \eta\frac{\partial L}{\partial b^{(l)}}$  





## Quy ước và công thức chung  

Với mạng có kiến trúc $n_0 - n_1 - n_2 - ... - n_L$  
Ta hiểu:  
* $n_0$: số nút đầu vào  
* $n_1$: số nơ-ron ở tầng l
* $n_L$: số nút đầu ra  
Nói cách khác $x_1,x_2 -> h_1, h_2 -> \hat{y}$  

<u><i>Kích thước các tham số ở tầng l:</i></u>  

$W^{(l)} \in R^{n_l \times n_{l-1}}, \quad b^{(l)} \in R^{n_l \times 1}$ 

<u><i>Kích thước các đại lượng trung gian:</i></u> 

$z^{(l)} \in R^{n_l \times 1}, \quad a^{(l)} \in R^{n_l \times 1}$  

<u><i>Số tham số ở tầng l:</i></u>   

$n_ln_{l-1} + n_l$  

<u><i>Tổng số tham số của mạng:</i></u> 

$\sum_{l=1}^L(n_ln_{l-1} + n_l)$  

<span style="color: red"> **Lưu ý** </span>  

- Bias **Không được tính** trong ký hiệu kiến trúc như 2-3-1  
- Mọi vector trong tài liệu được viết dưới dạng **vector cột**  
- Quy tắc nhớ nhanh:  
    * $W^{(l)}$  có số hàng bằng số nơ-ron tầng hiện tại.
    * $W^{(l)}$ có số cột bằng số nơ-ron tầng trước
